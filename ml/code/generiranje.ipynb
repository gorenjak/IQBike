```python
import pandas as pd
import numpy as np
from datetime import datetime, timedelta

# Ustvarimo prazen DataFrame za izposoje
rentals_df = pd.DataFrame(columns=['rental_id', 'start_date', 'end_date', 'start_station_id', 'end_station_id', 'bike_id', 'user', 'duration','trenutna_zasedenost_start','trenutna_zasedenost_end', 'weather'])

# Število izposoj, ki jih želimo generirati
n_rentals = 100

# Generiranje naključnih izposoj
for day in pd.date_range(start='2022-01-01', end='2022-12-31', freq='D'):
    datum = day.date()

    for i in range(n_rentals):
        zacetek = timedelta(hours=np.random.randint(0, 22), minutes=np.random.randint(1, 60), seconds=np.random.randint(1, 60))
        konec = zacetek + timedelta(minutes=np.random.randint(10, 50))
        rental_id = i
        start_date = datetime.combine(datum, datetime.min.time()) + zacetek
        end_date = datetime.combine(datum, datetime.min.time()) + konec
        start_station_id = np.random.choice(stations_df.index)

        # Check if kolesaArray is empty for start_station_id
        while len(stations_df.loc[start_station_id, 'kolesaArray']) == 0:
            start_station_id = np.random.choice(stations_df.index)

        end_station_id = np.random.choice(stations_df.index)

        # Check if kolesaArray has enough capacity for end_station_id
        while len(stations_df.loc[end_station_id, 'kolesaArray']) >= 21:
            end_station_id = np.random.choice(stations_df.index)

        bike_id = np.random.choice(stations_df.loc[start_station_id, 'kolesaArray'])
        stations_df.at[start_station_id, 'kolesaArray'].remove(bike_id)
        member_id = np.random.choice(users_df.index)
        member_type = 'Member'
        duration = end_date - start_date        
        weather = 'Sončno'
        #start_date = datetime.strptime(start_date, '%Y-%m-%d %H:%M:%S')

        # Update kolesaArray in end_station
        stations_df.at[end_station_id, 'kolesaArray'].append(bike_id)

        # Calculate the current occupancy of the end station
        trenutna_zasedenost_start = len(stations_df.loc[start_station_id, 'kolesaArray'])
        trenutna_zasedenost_end = len(stations_df.loc[end_station_id, 'kolesaArray'])

        rentals_df.loc[len(rentals_df)] = [rental_id, start_date, end_date, start_station_id, end_station_id, bike_id, member_id, duration, trenutna_zasedenost_start, trenutna_zasedenost_end, weather]
# Črpamo podatke iz DataFrame postajališča za start_station in end_station
rentals_df = rentals_df.merge(stations_df[['ime']], left_on='start_station_id', right_index=True)
rentals_df = rentals_df.rename(columns={'ime': 'start_station'})
rentals_df = rentals_df.merge(stations_df[['ime']], left_on='end_station_id', right_index=True)
rentals_df = rentals_df.rename(columns={'ime': 'end_station'})

# Prikaz podatkovne strukture za izposojo koles
ordered_df = rentals_df.sort_values(by='start_date', ascending=True)
ordered_df

```
