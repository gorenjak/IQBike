```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error

# Predpostavka: Imate DataFrame z imenom 'df', ki vsebuje značilke in ciljno spremenljivko, vključno s stolpcem 'station'.

# Izberite značilke in ciljno spremenljivko
features = ['temperature_2m (°C)', 'relativehumidity_2m (%)', 'precipitation (mm)',
       'weathercode (wmo code)', 'surface_pressure (hPa)', 'cloudcover (%)',
       'windspeed_10m (km/h)']
target = 'zasedenost'

# Ločite podatke na učni in testni nabor
X_train, X_test, y_train, y_test = train_test_split(df[features], df[target], test_size=0.2, random_state=42)

# Group by postaja
grouped = df.groupby('station')

# Napovedni model za vsako postajo
predictions = {}
for station, group in grouped:
    X = group[features]
    y = group[target]
    
    # Linearna regresija
    linear_reg = LinearRegression()
    linear_reg.fit(X, y)
    
    # Napovedi za trenutno postajo
    predictions[station] = linear_reg.predict(X_test[features])

# Izračun napake (MSE) za vsako postajo
mse = {}
for station, y_pred in predictions.items():
    mse[station] = mean_squared_error(y_test, y_pred)

# Izpis rezultatov
for station, mse_value in mse.items():
    print("Postaja:", station)
    print("MSE:", mse_value)
    print()
```

```python
zasedenost['date_only'] = zasedenost['date'].dt.date
datumi['date_only'] = datumi['date']

zasedenost['date_only'] = pd.to_datetime(zasedenost['date_only'])

# Merge the 'Tip' column from datumi_df into zasedenost_df based on matching dates
zasedenost = zasedenost.merge(datumi[['date_only', 'Tip']], on='date_only', how='left')

# Update 'Tip' in zasedenost_df where it is NaN with the value from datumi_df
zasedenost['Tip_x'].fillna(zasedenost['Tip_y'], inplace=True)

# Drop the 'Tip_y' column as it is no longer needed
zasedenost.drop('Tip_y', axis=1, inplace=True)

# Rename the 'Tip_x' column back to 'Tip'
zasedenost.rename({'Tip_x': 'Tip'}, axis=1, inplace=True)

zasedenost
```
